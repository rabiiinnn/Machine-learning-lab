```python
import numpy as np
```

```python
X = np.array([
[1, 1, 2], # Intercept, x1=1, x2=2
[1, 2, 1], # Intercept, x1=2, x2=1
[1, 3, 3] # Intercept, x1=3, x2=3
])
```

```python
y = np.array([6, 8, 12]).reshape(-1, 1)
```

```python
#OLS estimation
```

```python
#Transpose
XT = X.T
```

```python
#XT * X
XTX = XT@X
```

```python
#inverse
XTX_inv = np.linalg.inv(XTX)
```

```python
#calculate the
beta_hat =XTX_inv @ XT@y
```

```python
# Predicted values and residuals
y_hat = X @ beta_hat
residuals = y- y_hat
print("y_hat=",y_hat)
print("residuals",residuals)
```

```python
# Compute R-squared
SSR = np.sum(residuals ** 2) # Sum of Squared Residuals
SST = np.sum((y- np.mean(y)) ** 2) # Total Sum of Squares
R_squared = 1- SSR / SST
print(R_squared)
```

```python
# Compute Adjusted R-squared
n = len(y)          # number of observations
p = X.shape[1]      # number of parameters (including intercept)

if n > p:
    adj_R_squared = 1 - ((1 - R_squared) * (n - 1)) / (n - p)

else:
    adj_R_squared = None   # Not defined if degrees of freedom <= 0

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y.flatten(), y=y_hat.flatten())
plt.plot([min(y), max(y)], [min(y), max(y)], 'r--') # Red dashed line for y=y_hat
plt.xlabel("Actual Values (y)")
plt.ylabel("Predicted Values (y_hat)")
plt.title("Actual vs. Predicted Values")
plt.grid(True)
plt.show()
```
